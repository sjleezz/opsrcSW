```python
# 실습1

import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
max_result = []
min_result = []
for row in data:
    if row[-1] != '' and row[-2] != '':
        date = row[0].split('-')
        if date[1] == '09' and date[2] == '05' :
            max_result.append(float(row[-1]))
            min_result.append(float(row[-2]))
plt.plot(max_result) 
plt.plot(min_result)
plt.show()
```

```python
# 실습2

import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
jan = []
aug = []

for row in data :
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '01' :
            jan.append(float(row[-1]))
        elif month == '08' :
            aug.append(float(row[-1]))
plt.hist(jan, bins=100, color='b', label='Jan')
plt.hist(aug, bins=100, color='r', label='Aug')
plt.legend()
plt.show()
        
```
