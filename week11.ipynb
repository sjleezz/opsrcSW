```python
# 실습 8
class Calc:
    def sum(self, a, b) :
        result = a + b;
        print("{0} + {1} = {2}입니다".format(a, b, a+b)) 
    
    def sub(self, a, b) :
        result = a - b;
        print("{0} - {1} = {2}입니다".format(a, b, a-b)) 
        
    def mult(self, a, b) :
        result = a * b;
        print("{0} * {1} = {2}입니다".format(a, b, a*b)) 
        
    def divi(self, a, b) :
        result = a / b;
        print("{0} / {1} = {2}입니다".format(a, b, int(a/b))) 

def main() :
    calc = Calc()
    calc.sum(1,2)
    calc.sub(5,1)
    calc.mult(2,3)
    calc.divi(6,2)
    
    
if __name__ == '__main__' :
    main()
```

```python
# 실습 9
class Stack:
    #리스트를 이용하여 스택 생성
    def __init__ (self):
        self.top = []

    #스택 내부 자료를 string으로 변환하여 반환
    def __str__(self):
        return str(self.top[::1])

    #PUSH
    def push (self, item):
        self.top.append(item)

    #POP
    def pop(self):
        #if Stack is not empty
        if not self.isEmpty():
            #pop and return 
            return self.top.pop(-1)
        else:
            print("Stack underflow")
            exit()

    #스택 초기화
    def clear(self):
        self.top=[]


    #자료가 포함되어 있는지 여부 반환
    def isContain(self, item):
        return item in self.top
    
    
    #스택에서 top의 값을 읽어온다
    def peek(self):
        if not self.isEmpty():
            return self.top[-1]
        else:
            print("underflow")
            exit()

    #스택이 비어있는지 확인
    def isEmpty(self):
        return len(self.top)==0
    
    #스택 크기 반환
    def size(self):
        return len(self.top)
    
    
    
def checkOperatorTier(token) :
    if token == '*' or token == '/' :
        return 2
    elif token == '+' or token == '-' :
        return 1
    else :
        return 1

    
def calculate(NUM, OPT) :
   
    OPT_top = OPT.peek()   # 위에서부터 두개의 연산자를 top과 bottom으로 추출
    OPT.pop()
    OPT_bottom = OPT.peek()
    OPT.pop()
    
    top_tier = checkOperatorTier(OPT_top)       # 새로 들어온 연산자의 우선순위
    bottom_tier = checkOperatorTier(OPT_bottom) # 먼저 들어온 연산자의 우선순위
              
    if top_tier <= bottom_tier :      # 새로 들어온 연산자의 우선순위가 같거나 낮은 경우에 먼저 들어온 연산자를 실행
        NUM_top = NUM.peek()          # 위에서부터 두개의 피연산자를 top과 bottom으로 추출
        NUM.pop()
        NUM_bottom = NUM.peek()
        NUM.pop()
                
        result = ''
        if OPT_bottom == '+' :
            result = NUM_bottom + NUM_top
        elif OPT_bottom == '-' :
            result = NUM_bottom - NUM_top    
        elif OPT_bottom == '*' :
            result = NUM_bottom * NUM_top
        elif OPT_bottom == '/' :
            result = int(NUM_bottom / NUM_top)
                
        NUM.push(result)
        OPT.push(OPT_top)
        
        if OPT.size() > 1:
            calculate(NUM, OPT)
                
    else :
        OPT.push(OPT_bottom)
        OPT.push(OPT_top)


def main() :
    NUM = Stack()
    OPT = Stack()
    
    stmt = input('계산식 입력 : ')
    stmt = stmt +' $'    # 마지막 계산을 위한 더미 문자 --> 연산자 취급, 우선순위는 가장 낮다
    for tok in stmt.split() :
        if tok.isdigit() :         # 피연산자인 경우
            tok = int(tok)     
            NUM.push(tok)
            print('NUM : ', NUM)
        else :                     # 연산자인 경우
            OPT.push(tok)
            print('OPT : ', OPT)  
            
        if OPT.size() > 1:     # 연산자가 두 개 이상 쌓인 경우
            calculate(NUM, OPT)        
           
        
    print('계산식 출력 : ', NUM.peek())     # 결과 출력
    
    NUM.clear()
    OPT.clear()
            
if __name__ == '__main__' :
    main()
```
