```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv("./202012_202012_연령별인구현황_월간.csv", header=0, index_col=0, engine='python',encoding='CP949', thousands=',')
df1 = df.T
df1 = df1.sub(df1['서울특별시 광진구 (1121500000)'], axis=0)
df1 = np.power(df1, 2)
df_sum = df1.sum()
df_sum= df_sum.sort_values(ascending=False)
df_sum = df_sum[:-1]

name_far = str(df_sum.index[0])
name_far_short = name_far.split(' ')[1]

name_near = str(df_sum.index[-1])
name_near_short = name_near.split(' ')[1]

name_this = '서울특별시 광진구 (1121500000)'

print(name_far ,end=' : ')
print(df_sum.max())
print(name_near,end=' : ')
print(df_sum.min())

df_max = df.loc[name_far]
df_min = df.loc[name_near]
df_this = df.loc[name_this]

plt.figure(figsize=(14,8))
plt.style.use('ggplot')
df_max.plot()
df_min.plot()
df_this.plot()
plt.legend()
plt.show()
```
